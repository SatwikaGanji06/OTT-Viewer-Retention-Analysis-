```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Cognitive Load": [
        "Lower Cognitive Load",
        "Moderate Cognitive Load",
        "Higher Cognitive Load"
    ],
    "Avg Drop-off Probability": [
        0.40001434,
        0.528079148,
        0.659651696
    ]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(7,4))

plt.bar(
    df["Cognitive Load"],
    df["Avg Drop-off Probability"]
)

plt.title("Drop-off Probability Accelerates Beyond Cognitive Load Threshold")
plt.xlabel("Cognitive Load Bucket")
plt.ylabel("Average Drop-off Probability")

plt.tight_layout()
plt.show()
```
